**Median** <br>
Median memiliki pengertian sebagai nilai data tengah. Penyajian data yang dimaksud lebih kepada hasil dari penelitian. Biasanya data tersebut disajikan dalam bentuk tabel, diagram ataupun daftar dan kerap disebut sebagai statistik.

<center>Rumus Median :

Rumus Median dengan Jumlah data ganjil :

$$M_{e} = \frac{X (n + 1)}{2}$$

Rumus Median dengan Jumlah Data Genap :

$$M_{e} = \frac{X_\frac{n}{2} + X\left (\frac{n}{2}+1 \right )}{2}$$


```python
st.median(pH)
```

**Modus** <br>
Modus adalah data yang paling sering muncul. Modus merupakan ukuran pemusatan untuk menyatakan fenomena yang paling banyak terjadi. Sekumpulan data yang diperoleh, memungkinkan untuk memiliki nilai modus yang tidak tunggal atau mungkin juga tidak memilikinya.

```python
st.mode(pH)
```

**Mean** <br>
Mean adalah salah satu ukuran gejala pusat. Mean dapat dikatakan sebagai wakil kumpulan data. Menentukan mean dapat dilakukan dengan cara menjumlahkan seluruh nilai data, kemudian membaginya dengan banyaknya data.

<center> Rumus Mean :

$$mean = \frac{x_{1}+x_{2}+x_{3} ... x_{n}}{n}$$

```python
st.mean(pH)
```

Skewness <br>
Kemencengan atau kecondongan (skewness) adalah tingkat ketidaksimetrisan atau kejauhan simetri dari sebuah distribusi. Skewness diartikan sebagai kemiringan distribusi data.

```python
pH.skew()
```

**Equal Width Intervals** <br>
Equal-width intervals adalah discretization yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar.

**Equal Frequency Intervals** <br>
Equal-frequency intervals adalah discretization yang membagi data numerik menjadi beberapa kelompok dengan jumlah anggota yang kurang lebih sama besar.
